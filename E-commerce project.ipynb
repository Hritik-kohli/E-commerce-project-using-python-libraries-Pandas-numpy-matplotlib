***CONVERTING DATE COLUMNS***

```python
#syntax :data['Order Date']=pd.to_datetime(data['Order Date'])
     new column name ^    conversion^             ^  convert from
```

```python
data['Order Date']=pd.to_datetime(data['Order Date'])
data['Ship Date']=pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
# we need to add 3 new columns to our data to verify data monthly,weeklyand yearly

#SYNTAX:  data['Order Monthly']=data['Order Date'].dt.month
#   new column name^          Extract from^             ^ Extract what
```

```python
data['Order Monthly']=data['Order Date'].dt.month
data['Order Yearly']=data['Order Date'].dt.year
data['Order Day of week']=data['Order Date'].dt.dayofweek
```

```python
data.info()
```

# MONTHLY SALES ANALYSIS

```python
#Now we need to group by sales,month | sales,year | sales,week    to check sales month year and week-wise
# syntax: Sales_by_month = data.groupby('Order Monthly')['Sales'].sum().reset_index()
#                                                                        ^ Order Monthly--> its a index in this code and reset_index() will send 
#                                                                                            it back to data and treat it like a column                                     
```

```python
Sales_by_month = data.groupby('Order Monthly')['Sales'].sum().reset_index()
print(Sales_by_month)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(Sales_by_month['Order Monthly'], Sales_by_month['Sales'],color='purple', marker='o')
plt.title('Monthly Sales Analysis')       
plt.xlabel('Order Monthly')              
plt.ylabel('Sales')                      
plt.grid(True)
plt.tight_layout()
plt.show()                                 
```

# SALES AND PROFIT BY COUSTOMER SEGMENT

```python
Segmentwise_Sales_Profit = data.groupby("Segment")[['Sales', 'Profit']].sum().reset_index()

print(Segmentwise_Sales_Profit)
```

```python
Segmentwise_Sales_Profit["Sales_Profit"] = Segmentwise_Sales_Profit["Sales"] + Segmentwise_Sales_Profit["Profit"]
```

```python

plt.figure(figsize=(10, 6))
plt.pie(Segmentwise_Sales_Profit["Sales_Profit"],labels=Segmentwise_Sales_Profit["Segment"],autopct='%1.1f%%',startangle=140,colors=['lightgreen', 
'green','olive'] ,shadow=True)

plt.title('Segmentwise Sales + Profit Distribution')
plt.tight_layout()
plt.show()
```

**Consumer is highly commendable for our business**
